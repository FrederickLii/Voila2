One *perfect* answer is given by,

**Hall's marriage theorem** It is possible for all the girls to marry boys that they know if and only if for any group of girls, the group of boys they know (i.e., known by at least one girl) has a size bigger than or equal to the size of the group of girls.

With the language of graph theory,

**Hall's marriage theorem (for bipartite graphs)** Let $G = (X\sqcup Y,E)$ be a bipartite graph. Then $X$ can be matched to a subset of $Y$, if and only if, the neighborhood $|N(S)|$ is greater than or equal to $|S|$ for all subsets $S$ of $X$.

Here, a *matching* is a set of pairwise independent edges in this graph, i.e., no two edges in this collection share a same vertex. For a bipartite graph $(X\sqcup Y,E)$, $S\subseteq X$ can be matched to $T \subseteq Y$ means we have a matching $M$ with vertices $S \sqcup T$.

The "only if" part is trivial and we only care about the "if" part of this theorem.

**Proof 1.** Consider a *maximal* matching (namely, matching with the maximal number of edges) $M$ from $S \subseteq X$ to $T \subseteq Y$. Using $M$, we can assign directions for each edge in $G$ by
* $e = \{x,y\}$ goes from $x$ to $y$ if $e \notin M$
* $e = \{x,y\}$ goes from $y$ to $x$ if $e \in M$
 
Assume $S \neq X$, i.e., $\exists x \in X, x\notin S$. Consider the connect component $C_x$ of $x$.

* $C_x \bigcap Y = N(C_x \bigcap X)$, since any vertex $y\in Y$ adjunct to a vertex $x'\in X$ can either be reached by $x'$ or reach $x'$
* $|C_x \bigcap X| - 1 \geqslant |C_x \bigcap Y|$, since if $\exists v \in C_x \bigcap Y, v \notin T$, a path from $x$ to $v$ in $C_x$, say $xy_1x_1\cdots y_nx_n v$, we can find a matching $(M \setminus \{\{y_1,x_1\}, \cdots ,\{y_n,x_n\}\}) \sqcup \{\{x,y_1\},\cdots\{x_n,v\}\}$, which is bigger than $M$

However, $|C_x \bigcap X| \leqslant |N(|C_x \bigcap X|)|$. A contradiction.

<p style="text-align:right;">&#9632; </p>

This proof in fact provides an algorithm, which is the Ford–Fulkerson algorithm, whose input is a bipartite graph and output is a *maximal matching*. In cases where the Hall's conditions
$$
    |N(S)| \geqslant |S|, \forall S \subseteq X
$$
holds, the output will be a marriage arrangement, or say, a *perfect matching*. 

The pseudo-code is given as follows,

```python
# Ford–Fulkerson algorithm #########################
# INPUT : A bipartite graph (girls, boys, edges) ###
# Output: A maximal matching Matching ##############

Matching = set() # An empty set for matching edges #

for girl in girls:
    queue = Queue()
    queue.enqueue(girl)
    while queue is not Empty:
        this_girl = queue.dequeue()
        for boy in edges[this_girl]:
            if boy is in Matched:
                queue.enqueue(Matched[boy])
            else:
                Expand Matching by (girl,boy)
                break for
        if Matching has been expanded:
            break while

```

## Hall's marriage theorem as the simplest case of Kőnig's theorem

**Rethink 1** Since we have seen how to find a *maximal matching*, i.e. a matching with a maximal number of edges, for each bipartite graph $G = (X\sqcup Y,E)$, and such a maximal matching is obviously non-unique, we should think about the meaning of the **maximal value**, the size of a maximal matching, not the matching itself. Using this idea, let's rethink the meaning of Hall's marriage condition. Assume $|X| = n \leqslant |Y| = m$, $G$ has a perfect matching, i.e., a maximal matching of size $n$, if and only if for any $S \subseteq X$, $|S| + |Y| - |N(S)| \leqslant m$, which indicates that for any *independent set* $(S,T,\emptyset)$ as a induced subgraph of $G$, whose size $|S| + |T| \leqslant m$. Furthermore, the complement of an independent set is a *vertex cover* $(X\setminus S) \sqcup (Y \setminus T)$, namely, a subset of vertices such that for any edge $e \in E$, $e$ has at least one end in this subset. Therefore, the Hall's marriage theorem is equivalent to the following statement,

**Hall's marriage theorem (for bipartite graphs)** Let $G = (X\sqcup Y,E)$ be a bipartite graph. Then $X$ can be matched to a subset of $Y$, if and only if, $G$ is minimally covered by $X$.

or say,

**Hall's marriage theorem** It is possible for all the girls to marry boys that they know if and only if there is no  such a small group of boys and girls satisfies,
- If a girl knows a boy, then either the girl or the boy belongs to this group
- The size of this group is strictly smaller than the number of girls

This observation indeed show that Hall's marriage theorem is a special result of Kőnig's theorem,

**Kőnig's theorem** In any bipartite graph, the number of edges in a maximum matching equals the number of vertices in a minimum vertex cover. 

Hence, we have the second proof for Hall's marriage theorem,

**Proof 2.** Kőnig's theorem $\Longrightarrow$ Hall's marriage theorem.

<p style="text-align:right;">&#9632; </p>

**Remark.** These two theorems are in fact equivalent.

## Hopcroft–Karp–Karzanov algorithm

**Rethink 2.** If we calculate the time complexity of the above algorithm, we shall notice that the algorithm iterates over each vertex in the graph and then performs a BFS on the corresponding edges to find an *augmenting path* (i.e., a path starting from a unmatched vertex and along unmatched edges and matched edges alternatively and ending at an unmatched vertex), which in the worst cases, needs $\mathcal{O}(|E|) \times \mathcal{O}(|V|)$ time to run, or say, the worst time complexity is $\mathcal{O}(|E||V|)$. However, one can make this process quicker, by searching all a maximal set of shortest augmenting paths, we can expand the matching more rapidly. The pseudo-code is given as follows,

```python
# Hopcroft–Karp–Karzanov algorithm algorithm #######
# INPUT : A bipartite graph G = (U,V,edges) ########
# Output: The size of maximal Matching #############

# Prepare an additional vertex to match ############
# all unmatched vertices ###########################
NIL = Vertex()

# Prepare two arrays for recording matchings #######
# for unmatched vertices assign NIL ################
Pair_U = [NIL] * length(U)
Pair_V = [NIL] * length(V)

# In each phase of iterations we use BFS and DFS ###
# to find augmenting paths of the shortest length ##
# Starting from all unmatched vertices in U ########
# which is regarded as a layer of depth 0 ##########
# we shall split U into layers by distance #########
# from layer 0 by shortest augmenting paths ########
# Prepare an array to record the distance ##########
# which will be reset before each phase ############
Dist = [INF] * (length(U) + 1)

# To find the least length of augmenting paths #####
# we use BFS to find it and record into Dist #######
# The existence of such paths is equivalent ########
# to Dist[NIL] < INF, thus NIL can be reached ######
# Return if we can find one ########################
def BFS():
    Q = Queue()
    for u in U:
        if Pair_U[u] == NIL: # if u is unmatched ###
            Dist[u] = 0 # put u into the 0th layer #
            Q.enqueue(u)
        else:
            Dist[u] = INF # reassign the distance ##
    Dist[NIL] = INF
    while Q is not Empty:
        u == Q.dequeue()
        # only search shortest augmenting paths ####
        if Dist[u] < Dist[NIL]:
            for v in edges[u]:
                if Dist[Pair_V[v]] == INF:
                    Dist[Pair_V[v]] = Dist[u] + 1
                    Q.enqueue(Pair_V[v])
    return Dist[NIL] != INF

# We then use DFS from each layer 0 vertex #########
# to find a maximal set of all augmenting paths ####
# of the shortest length to expand the matching ####
# Return if we can find one ########################
def DFS(u):
    if u != NIL:
        for v in edges[u]:
            if Dist[Pair_V[v]] == Dist[u] + 1:
                if DFS(Pair_V[v]):
                    Pair_V[v] = u
                    Pair_U[u] = v
                    return True
        Dist[u] = INF
        return False
    return True

matching = 0
while BFS():
    for u in U:
        if Pair_U[u] == NIL:
            if DFS(u):
                matching = matching + 1
```

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["animation.html"] = "jshtml"
import matplotlib.animation
import numpy as np
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import random
import matplotlib.patches as patch
from IPython.display import display,clear_output
from celluloid import Camera
from collections import defaultdict


output_widget2 = widgets.Output()

class BipartiteGraph(object):
    def __init__(self, data):
        m = len(data)
        n = max(max([-1] + u) for u in data) + 1
        self.__NIL = m
        self.__INF = float('INF')
        self.Pair_U = [m] * m
        self.Pair_V = [m] * n
        self.Dist = [self.__INF] * (m + 1)
        self.edges = data
        self.m,self.n = m,n
        self.size = max(m,n)
        self.path = defaultdict(list)
        m = self.size
        self.fig,self.ax = plt.subplots()
        self.camera = Camera(self.fig)
        
        self.ax.set_axis_off()

        self.ax.set_xlim(-1.3,1.7)
        self.ax.set_ylim(-.5,m + 1.5)
        self.ax.set_aspect('equal')
        self.hint = {'x': -1.3, 'y': m + 0.5, 'text':'The bipartite graph is drawn!'}
        self.colors = {
            0:'#FEF9E7', 
            1:'#F9E79F', 
            2:'#F4D03F', 
            3:'#F1C40F', 
            4:'#D4AC0D', 
            5:'#B7950B', 
            6:'#9A7D0A', 
            self.__INF:'#7D6608'}
        
        
    def __str__(self):
        return str([self.__NIL,
        self.__INF ,
        self.Pair_U,
        self.Pair_V,
        self.Dist,
        self.edges,
        self.path])
    
          
    def BFS(self):
        
        self.hint['text'] = 'Run BFS to find a layer decomposition of U.'
        self.draw()
        Q = []
        for u in range(self.m):
            if self.Pair_U[u] == self.__NIL:
                self.Dist[u] = 0 
                Q.append(u)
            else:
                self.Dist[u] = self.__INF
            self.Dist[self.__NIL] = self.__INF
        if Q:
            temp = ','.join(['U[' + str(q) + ']' for q in Q])
            self.hint['text'] = f'Layer 0 contains {temp}.'
        else:
            self.hint['text'] = 'No unmatched vertex is left.'
        self.draw()
        while Q:
            u = Q.pop(0)
            if self.Dist[u] < self.Dist[self.__NIL]:
                for v in self.edges[u]:
                    if self.Dist[self.Pair_V[v]] == self.__INF:
                        self.Dist[self.Pair_V[v]] = self.Dist[u] + 1
                        Q.append(self.Pair_V[v])
                        if self.Pair_V[v] == self.__NIL:
                            self.hint['text'] = f'The least length of augmenting paths is {self.Dist[self.Pair_V[v]] * 2 - 1},\nStop BFS.'
                            self.draw()
                        else:
                            self.hint['text'] = f'Vertex U[{self.Pair_V[v]}] is put into layer {self.Dist[self.Pair_V[v]]}.'
                            self.draw()
        if self.Dist[self.__NIL] != self.__INF:
            return True
        else:
            self.hint['text'] = 'No augmenting path is found.\nStop BFS.'
            self.draw()
            return False

    def DFS(self, u):
        if u !=self.__NIL:
            for v in self.edges[u]:
                if self.Dist[self.Pair_V[v]] ==self.Dist[u] + 1:
                    if self.DFS(self.Pair_V[v]):
                        if self.Pair_V[v] != self.__NIL:
                            self.path[self.Pair_V[v]].append(v)
                        self.path[u].append(v)
                        self.Pair_V[v] = u
                        self.Pair_U[u] = v
                        return True
            self.Dist[u] =self.__INF
            return False
        return True
    
    def draw(self):
        self.ax.text(
            -1.3, self.size+1,
            'Hopcroft–Karp–Karzanov algorithm', 
            fontsize = 11,
            weight='bold')

        x,y,text = self.hint['x'],self.hint['y'],self.hint['text']
        self.ax.text(x,y,text,fontsize = 10,va = 'top')
        
        for g in range(self.m):
            for b in self.edges[g]:
                x0,y0 = -0.8,g + 0.1
                line, = self.ax.plot(np.array([x0,1.2]), np.array([y0,b+0.1]),color = '#808080',zorder = 0)
            if self.Pair_U[g] != self.__NIL:
                b = self.Pair_U[g]
                x0,y0 = -0.8,g + 0.1
                line, = self.ax.plot(np.array([x0,1.2]), np.array([y0,b+0.1]),color = '#FF0000',zorder = 0)
                
        if self.path:
            for u in self.path:
                for v in self.path[u]:
                    
                    x0,y0 = -0.8, u + 0.1
                    line, = self.ax.plot(np.array([x0,1.2]), np.array([y0,v+0.1]),color = 'b',zorder = 0)
                
        u_boxes = {}
        for g in range(self.m):
            col = self.colors[self.Dist[g]]
            box = patch.FancyBboxPatch(
                (-1, g), 0.4, 0.2,
                boxstyle=patch.BoxStyle("Round", pad=0.2),
                color = col,
                zorder = 5
            )
            self.ax.add_artist(box)
            text = self.ax.text(
                -0.8, 0.1+g,
                f'U[{g}]', 
                color='k', 
                style = 'italic',
                weight='bold', 
                fontsize=12, 
                ha='center', 
                va='center',
                zorder = 10)
            u_boxes[g] = {'box':box,'text':text}
        
        v_boxes = {}
        c = 0
        for b in range(self.n):
            box = patch.FancyBboxPatch(
                (1, b), 0.4, 0.2,
                boxstyle=patch.BoxStyle("Round", pad=0.2),
                color = '#0066CC',
                zorder = 5
            )
            self.ax.add_artist(box)
            text = self.ax.text(
                1.2, 0.1+b,
                f'V[{b}]', 
                color='k', 
                style = 'italic',
                weight='bold', 
                fontsize=12, 
                ha='center', 
                va='center',
                zorder = 10
            )
            v_boxes[b] = {'box':box,'text':text,'x':1.2,'y':b+0.1}
        self.camera.snap()

    def HKK(G):
        matching = 0
        while G.BFS():
            G.draw()
            for u in range(G.m):
                
                if G.Pair_U[u] == G.__NIL:
                    G.path = defaultdict(list)
                    G.hint['text'] = f'Run DFS to find an augmenting path starting from U[{u}].'
                    G.draw()
                    if G.DFS(u):
                        G.hint['text'] = f'An augmenting path starting from U[{u}] is found.\nStop DFS.'
                        G.draw()
                        G.path = defaultdict(list)
                        G.hint['text'] = 'Expand the matching by this path.'
                        G.draw()
                        matching = matching + 1
                    else:
                        G.hint['text'] = f'No augmenting path starting from U[{u}] is found.\nStop DFS.'
                        G.draw()
                        G.path = defaultdict(list)

        G.hint['text'] = f'The Hopcroft–Karp–Karzanov algorithm stops.\nThe maximal matching has {matching} pairs.'
        G.draw()
        G.draw()
        ani = G.camera.animate(interval = 1000)
        with output_widget2:
            clear_output()
            display(ani)

input_text2 = widgets.Text(
    value='',
    placeholder='Input a 2D list like [[0,],[2,],[1,3,]]',
    description='Input array',
    disabled=False,  
    continuous_update= False
)

display(input_text2)

def HKK_widget(change):
    a = change['new']
    with output_widget2:
        
        clear_output()
        
        # check if the input is a 2D list ###
        if not a:
            print('Input is empty, try to input a 2D list')
            return
        try:
            data = eval(a)
        except:
            print("Illegal inputs")
            return
        try:
            if type(data) != list:
                print("Input should be like [[0,],[2,],[1,3,]]")
                return
            if not data:
                print('Input is empty, try to input a 2D list')
                return
            for girl in data:
                if type(girl) != list:
                
                    print("Input should be like [[0,],[2,],[1,3,]]")
                    for boy in girl:
                        if type(boy) != int:
                            raise Exception("Illegal inputs")
                    return
        except:
            print("Illegal inputs")
            return
    G = BipartiteGraph(data) 
    G.HKK()
    
input_text2.observe(HKK_widget, names = 'value')
display(output_widget2)
````
