One *perfect* answer is given by,

**Hall's marriage theorem** It is possible for all the girls to marry boys that they know if and only if for any group of girls, the group of boys they know (i.e., known by at least one girl) has a size bigger than or equal to the size of the group of girls.

With the language of graph theory,

**Hall's marriage theorem (for bipartite graphs)** Let $G = (X\sqcup Y,E)$ be a bipartite graph. Then $X$ can be matched to a subset of $Y$, if and only if, the neighborhood $|N(S)|$ is greater than or equal to $|S|$ for all subsets $S$ of $X$.

Here, a *matching* is a set of pairwise independent edges in this graph, i.e., no two edges in this collection share a same vertex. For a bipartite graph $(X\sqcup Y,E)$, $S\subseteq X$ can be matched to $T \subseteq Y$ means we have a matching $M$ with vertices $S \sqcup T$.

The "only if" part is trivial and we only care about the "if" part of this theorem.

**Proof 1.** Consider a *maximal* matching (namely, matching with the maximal number of edges) $M$ from $S \subseteq X$ to $T \subseteq Y$. Using $M$, we can assign directions for each edge in $G$ by
* $e = \{x,y\}$ goes from $x$ to $y$ if $e \notin M$
* $e = \{x,y\}$ goes from $y$ to $x$ if $e \in M$
 
Assume $S \neq X$, i.e., $\exists x \in X, x\notin S$. Consider the connect component $C_x$ of $x$.

* $C_x \bigcap Y = N(C_x \bigcap X)$, since any vertex $y\in Y$ adjunct to a vertex $x'\in X$ can either be reached by $x'$ or reach $x'$
* $|C_x \bigcap X| - 1 \geqslant |C_x \bigcap Y|$, since if $\exists v \in C_x \bigcap Y, v \notin T$, a path from $x$ to $v$ in $C_x$, say $xy_1x_1\cdots y_nx_n v$, we can find a matching $(M \setminus \{\{y_1,x_1\}, \cdots ,\{y_n,x_n\}\}) \sqcup \{\{x,y_1\},\cdots\{x_n,v\}\}$, which is bigger than $M$

However, $|C_x \bigcap X| \leqslant |N(|C_x \bigcap X|)|$. A contradiction.

<p style="text-align:right;">&#9632; </p>

This proof in fact provide an algorithm, which is the Hopcroft–Karp–Karzanov algorithm, whose input is a bipartite graph and output is a maximal matching. In cases where the Hall's conditions
$$
    |N(S)| \geqslant |S|, \forall S \subseteq X
$$
holds, the output will be a marriage arrangement, or say, a *perfect matching*. 

The pseudo-code is given as follows,

```python
# Hopcroft–Karp–Karzanov algorithm #################
# INPUT : A bipartite graph (girls, boys, edges) ###
# Output: A maximal matching Matching ##############

Matching = set() # An empty set for matching edges #

for girl in girls:
    queue = Queue()
    queue.enqueue(girl)
    while queue is not Empty:
        this_girl = queue.dequeue()
        for boy in edges[this_girl]:
            if boy is in Matched:
                queue.enqueue(Matched[boy])
            else:
                Expand Matching by (girl,boy)
                break for
        if Matching has been expanded:
            break while

```

Input a 2D list ```girls``` in the following widget, you will see how this algorithm works. Each item ```girls[k]``` in ```girls``` should be a ```list``` of several integers to represent the boys girl $k$ knows. For example, ```[[0,],[2,],[1,3,]]``` indicates that
- girl $0$ knows boy $0$
- girl $1$ knows boy $2$
- girl $2$ knows boy $1$ and boy $3$

````python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patch
from IPython.display import display,clear_output
import time

input_text = widgets.Text(
    value='',
    placeholder='Input a 2D list like [[0,],[2,],[1,3,]]',
    description='Input array',
    disabled=False,  
    continuous_update= False
)


output_widget = widgets.Output()

display(input_text)

def HKK(change):
    a = change['new']
    with output_widget:
        
        clear_output()
        # check if the input is a 2D list ###
        if not a:
            print('Input is empty, try to input a 2D list')
            return
        try:
            girls = eval(a)
        except:
            print("Illegal inputs")
            return
        try:
            if type(girls) != list:
                print("Input should be like [[0,],[2,],[1,3,]]")
                return
            for girl in girls:
                if type(girl) != list:
                
                    print("Input should be like [[0,],[2,],[1,3,]]")
                    for boy in girl:
                        if type(boy) != int:
                            raise Exception("Illegal inputs")
                    return
        except:
            print("Illegal inputs")
            return
        # create a hash map to record boys ##
        boy_set = {}
        for g in range(len(girls)):
            girl = girls[g]
            for boy in girl:
                if boy not in boy_set:
                    boy_set[boy]=[]
                boy_set[boy].append(g)
        # prepare the girls and boys nodes ##
        m = max(len(boy_set),len(girls),1)
        fig = plt.figure(figsize=(4, (m+1)/2), layout="constrained")
        axs = fig.subplots(1, 2)
        ax = axs[0]
        ax.set_xlim(-1.5,2.5)
        ax.set_ylim(-.5,m-0.5)
        ax.set_aspect('equal')
        ax.set_axis_off()
        c = 0
        for b in boy_set:
            for g in boy_set[b]:
                x0,y0 = -0.8,g + 0.1
                line, = ax.plot(np.array([x0,1.2]), np.array([y0,c+0.1]),color = '#808080',zorder = 0)
            c += 1
            
        # set boxes for girls ######
        for g in range(len(girls)):
            box = patch.FancyBboxPatch(
                (-1, g), 0.4, 0.2,
                boxstyle=patch.BoxStyle("Round", pad=0.2),
                color = '#FFD700',
                zorder = 5
            )
            ax.add_artist(box)
            text = ax.text(
                -0.8, 0.1+g,
                f'Girl {g}', 
                color='w', 
                style = 'italic',
                weight='bold', 
                fontsize=7, 
                ha='center', 
                va='center',
                zorder = 10)
        c = 0
        
        # set boxes for boys ######
        for b in boy_set:
            box = patch.FancyBboxPatch(
                (1, c), 0.4, 0.2,
                boxstyle=patch.BoxStyle("Round", pad=0.2),
                color = '#0066CC',
                zorder = 5
            )
            ax.add_artist(box)
            text = ax.text(
                1.2, 0.1+c,
                f'Boy {b}', 
                color='w', 
                style = 'italic',
                weight='bold', 
                fontsize=7, 
                ha='center', 
                va='center',
                zorder = 10
            )
            c += 1
        ax = axs[1]
        ax.set_xlim(-1.5,2.5)
        ax.set_ylim(-.5,m-0.5)
        ax.set_aspect('equal')
        ax.set_axis_off()
        c = 0
        lines = {}
        for b in boy_set:
            for g in boy_set[b]:
                x0,y0 = -0.8,g + 0.1
                line, = ax.plot(np.array([x0,1.2]), np.array([y0,c+0.1]),color = '#808080',zorder = 0)
                lines[(g,b)] = line
            c += 1
            
        girls_box = {}
        for g in range(len(girls)):
            box = patch.FancyBboxPatch(
                (-1, g), 0.4, 0.2,
                boxstyle=patch.BoxStyle("Round", pad=0.2),
                color = '#FFD700',
                zorder = 5
            )
            ax.add_artist(box)
            text = ax.text(
                -0.8, 0.1+g,
                f'Girl {g}', 
                color='w', 
                style = 'italic',
                weight='bold', 
                fontsize=7, 
                ha='center', 
                va='center',
                zorder = 10)
            
            girls_box[g] = {'box':box,'text':text,'x':-0.8,'y':g+0.1}
        boys_box = {}
        c = 0
        for b in boy_set:
            box = patch.FancyBboxPatch(
                (1, c), 0.4, 0.2,
                boxstyle=patch.BoxStyle("Round", pad=0.2),
                color = '#0066CC',
                zorder = 5
            )
            ax.add_artist(box)
            text = ax.text(
                1.2, 0.1+c,
                f'Boy {b}', 
                color='w', 
                style = 'italic',
                weight='bold', 
                fontsize=7, 
                ha='center', 
                va='center',
                zorder = 10
            )
            boys_box[b] = {'box':box,'text':text,'x':1.2,'y':c+0.1}
            c += 1
        fig.suptitle(f'Try to match {len(girls)} girls')
        display(fig)
        time.sleep(2)
        
        girls_matched = {}
        boys_matched = {}
        for g0 in range(len(girls)):
            girls_visited = {}
            boys_visited = {}
            stack = [g0]
            flag = True
            while stack and flag:
                g = stack.pop(0)    
                girls_box[g]['box'].set_color('k')
                fig.suptitle(f'Check all the boys girl {g} knows')
                clear_output()
                display(fig)
                time.sleep(1)
                for b in girls[g]:
                    if b not in boys_visited:
                        if b not in boys_matched:
                            fig.suptitle(f'Boy {b} has NOT been matched')
                            flag = False
                            boys_matched[b] = g
                            girls_matched[g] = b
                            lines[(g,b)].set_color('#FF0000')
                            while g in girls_visited:
                                b = girls_visited[g]
                                lines[(g,b)].set_color('#808080')
                                g = boys_visited[b]
                                boys_matched[b] = g
                                girls_matched[g] = b
                                lines[(g,b)].set_color('#FF0000')
                            break
                        else:
                            boys_box[b]['box'].set_color('k')
                            fig.suptitle(f'Boy {b} has been matched')
                            clear_output()
                            display(fig)
                            time.sleep(1)
                            newg = boys_matched[b]
                            stack.append(newg)
                            boys_visited[b] = g
                            girls_visited[newg] = b
                clear_output()
                display(fig)
                time.sleep(1)
            if not flag:
                fig.suptitle(f'Expand matching!')
            else:
                fig.suptitle(f'No unmatched boy is found!')
                
            for b in boys_visited:
                boys_box[b]['box'].set_color('#0066CC')
            for g in girls_visited:
                girls_box[g]['box'].set_color('#FFD700')
            girls_box[g0]['box'].set_color('#FFD700')
            clear_output()
            display(fig)
            time.sleep(1)
        if len(girls_matched) == len(girls):
            fig.suptitle(f"Hall's marriage condition is satisfies!")
        else:
            fig.suptitle(f"Hall's marriage condition is NOT satisfies!")
        clear_output()
        display(fig)
        time.sleep(1)

input_text.observe(HKK, names = 'value')
display(output_widget)



````
