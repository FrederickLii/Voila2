# Matchings and partitions

## Matchings, non-nesting matchings and non-crossing matchings
A *matching* on a totally ordered set of $2n$ elements $\{1,\cdots,2n\}$ is an unordered sequence of $n$ chords $\{c_i = (s_i,t_i)\}$ with $s_i < t_i$, such that $\{s_i\} \bigsqcup \{t_i\} = \{1,\cdots, 2n\}$. In other words, we splits $2n$ elements into two groups, the source group and the target group, by matching elements from different groups into pairs ```(source[i],target[i])``` or simply, $(s_i,t_i)$ under the condition that the source is always smaller than the target.

There are two special kinds of matchings,
* *non-crossing matchings* or *purely-nesting matchings*, which satisfy that $\forall 1 \leqslant i \neq j \leqslant n$, $s_i < s_j < t_i$ if and only if $s_i < t_j < t_i$
* *non-nesting matchings* or *purely-crossing matchings*, which satisfy that $\forall 1 \leqslant i \neq j \leqslant n$, $s_i < s_j < t_i$ if and only if $ s_j < t_i < t_j$

It is not hard to see that
* A non-crossing matching of $2n$ elements is exactly the same as a Dyck code regarded as a sequence of ```'push'``` and ```'pop'``` of a stack, or equivalently, an allowable sequences of parenthesizes
* A non-nesting matching of $2n$ elements is exactly the same as a Dyck code regarded as a sequence of ```'enqueue'``` and ```'dequeue'``` of a queue.

The duality between non-crossing objects and non-nesting objects is a key phenomenon in the world of catalan numbers, which indicates that for once we have a bijection from the set of all Dyck codes to some other "space", we in fact has two totally different bijections, one carries the stack structure while the rest one carries the queue structure in the constructions for algorithms.

## Partitions, non-crossing partitions and non-nesting partitions

A *partition* on a totally ordered set $X = \{1,\cdots,n\}$ is an equivalence relations on $X$, a.k.a, a decomposition of $X$ into an unordered family of subset $\{U_i\subseteq X\}$, such that
* $U_i \neq \emptyset$
* $X = \bigsqcup\limits_i U_i$

However, one need to notice that $X$ carries a total order and so does each equivalence class $U_i$. Hence we can describe a partition clearly if we know for each element $k \in U_i$ if $k$ has a successor and if so, which one the successor is. Using this idea, we associate each partition with a collections of tuples ```(prev[i],next[i)``` or simply, $\{(p_i,n_i\}$ to mark that the previous one $p_i$ has a successor $n_i$ as the next one, such that
* $p_i < n_i$
* $\forall i \neq j, p_i \neq p_j$ and $n_i \neq n_j$

Using this idea, we can use a similar idea to define two specific types of partitions, just as what we have done for matchings, *non-nesting partitions* and *non-crossing partitions*. 

* *non-crossing partitions* or *purely-nesting partitions*, which satisfy that $\forall i \neq j $, $p_i < p_j < n_i$ if and only if $p_i < n_j < n_i$
* *non-nesting partitions* or *purely-crossing partitions*, which satisfy that $\forall i \neq j $, $p_i < p_j < n_i$ if and only if $ p_j < n_i < n_j$

## How to get non-crossing/non-nesting partitions from non-crossing/non-nesting matchings?

Now, assume we have a non-crossing or non-nesting matching of $2n$ elements, which contains $n$ chords $(s_i,t_i)$. To make discussion simpler, we further assume $\{s_i\}$ is increasing, i.e., $1 = s_1 < s_2 < \cdots < s_n$. The idea for these constructions are very simple:
- **Step 1.** We collect all the chords $c_i = (s_i,t_i)$ to get a totally ordered set $c_1 < \cdots < c_n$. The order is given by the first components, i.e., $c_i < c_j \iff s_i < s_j \iff i < j$. 
- **Step 2.** For a chord $c_i$, we say it has a successor $c_j$ if and only if $t_i +1 = s_j$ and then we split $n$ chords into classes with this relation. In other words,
    - $c_i$ is the first element of some class if and only if $s_{i-1} = s_i - 1$
    - $c_i$ is the last element of some class if and only if $t_i + 1 \neq s_j$ for any $i < j \leqslant n$.
    - $\{c_{i_1} < c_{i_2} \cdots < c_{i_k}\}$ is a class if and only if $c_{i_1}$ satisfies the first condition above and $c_{i_k}$ satisfies the second one and for any $0<j<k$, $t_{i_j} + 1 = s_{i_{j+1}}$

Although one can check this construction works for any matchings, we restrict the discussion in the cases either non-crossing or non-nesting condition is guaranteed, since without these conditions we can't recover matchings uniquely from partitions and, as we shall see later, the construction above preserves these conditions well.

### Non-crossing cases
Starting with a non-crossing matching, the partition constructed as above satisfies, given two successor tuples $(p_1,n_1) = (c_i,c_j)$ and $(p_2,n_2) = (c_k,c_l)$,
$$
    c_i < c_k < c_j \iff s_i < s_k < s_j \iff s_i < s_k < t_i \iff s_i < t_k < t_i \iff s_i < s_l < t_i \iff s_i < s_l < s_j \iff c_i < c_l < c_j. 
$$
Hence, the constructed partition is still non-crossing.
