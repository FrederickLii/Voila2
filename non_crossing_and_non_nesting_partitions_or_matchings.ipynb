# Matchings and partitions

## Matchings, non-nesting matchings and non-crossing matchings
A *matching* on a totally ordered set of $2n$ elements $\{1,\cdots,2n\}$ is an unordered squence of $n$ chords $\{c_i = (s_i,t_i)\}$ with $s_i < t_i$, such that $\{s_i\} \bigsqcup \{t_i\} = \{1,\cdots, 2n\}$. In other words, we splits $2n$ elements into two groups, the source group and the target group, by matching elements from different groups into pairs under the condition that the source is always smaller than the target.

There are two special kinds of matchings,
* *non-crossing matchings* or *purely-nesting matchings*, which satifies that $\forall 1 \leqslant i \neq j \leqslant n$, $s_i < s_j < t_i$ if and only if $s_i < t_j < t_i$
* *non-nesting matchings* or *purely-crossing matchings*, which satifies that $\forall 1 \leqslant i \neq j \leqslant n$, $s_i < s_j < t_i$ if and only if $ s_j < t_i < t_j$

It is not hard to see that
* A non-crossing matching of $2n$ elements is exactly the same as a Dyck code regarded as a sequence of ```'push'``` and ```'pop'``` of a stack, or equivalently, an allowable squences of parenthesizes
* A non-nesting matching of $2n$ elements is exactly the same as a Dyck code regarded as a sequence of ```'enqueue'``` and ```'dequeue'``` of a queue.

The duality between non-crossing objects and non-nesting objects is a key phenomenon in the world of catalan numbers, which indicates that for once we have a bijection from the set of all Dyck codes to some other "space", we in fact has two totally different bijections, one carries the stack structure while the rest carries the queue structure in the constructions for algorithms.  

````python
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patch
import matplotlib.path as mpath
from IPython.display import display,clear_output
import ipywidgets as widgets
import time
from collections import Counter

class DyckPath:
    def __init__(self, path : np.ndarray = np.array([])):
        self.path = path

    def __str__(self):
        return str(self.path)
    
    def toCode(self):
        last = 0
        code = []
        for pos in self.path:
            code.append(pos-last)
            last = pos
        return DyckCode(np.array(code))


class DyckCode:
    def __init__(self, code : np.ndarray = np.array([])):
        self.code = code

    def __str__(self):
        return str(self.code)

    def toPath(self):
        last = 0
        path = []
        for move in self.code:
            path.append(move+last)
            last = path[-1]
        return DyckPath(np.array(path))

    @staticmethod
    def randomCode(length = 0):
        code = []
        pos,neg,check = 0,0,0
        for _ in range(length << 1):
            if check == 0:
                seed = 1
                check += 1
            else:
                seed = random.randint(0,1)
                check += (seed << 1) - 1
            if seed:
                pos += 1
                code.append(1)
                if pos == length:
                    code = code + [-1] * (length - neg)
                    break
            else:
                neg += 1
                code.append(-1)
                if neg == length:
                    code = code + [1] * (length - pos)
        return DyckCode(np.array(code))
    
def generatingMatchings(change):
    l = random.randint(1,10)
    code = DyckCode.randomCode(l)
    path = code.toPath()
    dyckCode_UR = ''.join([{1:'E',-1:'D'}[d] for d in code.code])
    with output_widget:
        fig = plt.figure(figsize=(6, 2), layout="constrained")
        axs = fig.subplots(1, 2)
        for ax in axs:
            ax.set_xlim(-2,2*l+2)
            ax.set_ylim(-5,5)
        
        plt.title(f"For Dyck code of length {2*l}: {dyckCode_UR}")
        axs[0].set_aspect('equal')
        axs[1].set_aspect('equal')
        axs[1].set_axis_off()
        axs[0].set_axis_off()
        axs[0].plot(np.arange(0,2*l+2,1),np.array([0]*(2*l+2)),'b')
        axs[1].plot(np.arange(0,2*l+2,1),np.array([0]*(2*l+2)),'b')
        
        clear_output(wait=True)
        display(fig)
        time.sleep(2)
        stack = []
        queue = []
        for x in range(1,2*l + 1):
            axs[0].set_title(dyckCode_UR[:x],y=0)
            if dyckCode_UR[x-1] == 'E':
                stack.append(x)
                queue.append(x)
                axs[0].plot(x,0,'bo')
                axs[0].text(x-0.1,-1,x)
                axs[1].text(x-0.1,-1,x)
                axs[1].plot(x,0,'bo')
            else:
                y = stack.pop()
                z = queue.pop(0)
                axs[0].plot(x,0,'ro')
                axs[0].text(x-0.1,-1,x)
                axs[1].text(x-0.1,-1,x)
                axs[1].plot(x,0,'ro')
                Path = mpath.Path
                pp1 = patch.PathPatch(Path([(y, 0.6), ((x+y)/2, (x-y) + 1), (x, 0.6)],[Path.MOVETO, Path.CURVE3, Path.CURVE3]),fc="none", transform=axs[0].transData)
                axs[0].add_patch(pp1)
                pp1 = patch.PathPatch(Path([(z, 0.6), ((x+z)/2, (x-z) + 1), (x, 0.6)],[Path.MOVETO, Path.CURVE3, Path.CURVE3]),fc="none", transform=axs[1].transData)
                axs[1].add_patch(pp1)
            axs[1].set_title(dyckCode_UR[:x],y=0)
            clear_output(wait=True)
            display(fig)
            time.sleep(1)
            
        
    


button = widgets.Button(description="Random Dyck code")
output_widget = widgets.Output()

display(button)

button.on_click(generatingMatchings)
display(output_widget)


        
````
